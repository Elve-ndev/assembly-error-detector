```python
import os, re, pickle, glob
import numpy as np
import pandas as pd
import torch
import torch.nn as nn
import torch.nn.functional as F
from collections import defaultdict
from sklearn.metrics import roc_auc_score, roc_curve
from scipy.spatial.distance import mahalanobis
from numpy.linalg import pinv
import matplotlib.pyplot as plt

# ── Chemins ───────────────────────────────────────────────
TRAIN_FEAT_DIR = '/kaggle/input/datasets/hibabou/industrial/data/train/'
VAL_FEAT_DIR   = '/kaggle/input/datasets/hibabou/industrial/data/val/'
TEST_FEAT_DIR  = '/kaggle/input/datasets/hibabou/featuressupp/sf_features_test/'
PREP_DIR       = '/kaggle/input/datasets/hibabou/prepbigru/kaggle/working/'
CKPT_PATH      = '/kaggle/input/datasets/hibabou/checkpoint0-66/bigru_2classes_best.pth'
PSR_TRAIN_DIR  = '/kaggle/input/datasets/hibabou/psr-labels/kaggle/working/psr_labels/'
PSR_VAL_DIR    = '/kaggle/input/datasets/hibabou/psr-labels-val/kaggle/working/psr_labels_val/'
DEVICE         = torch.device('cpu')
FEAT_DIM       = 2304
NUM_CLASSES    = 2

# ── Semantic groups ───────────────────────────────────────
SEMANTIC_GROUPS_2 = {
    0:0, 2:0, 4:0, 5:0, 8:0, 9:0, 10:0, 11:0, 13:0,
    15:0, 17:0, 18:0, 20:0, 21:0, 23:0, 27:0, 43:0, 65:0,
    12:0, 14:0, 24:0, 35:0, 36:0, 40:0, 42:0, 44:0, 45:0,
    47:0, 48:0, 52:0, 54:0, 57:0, 59:0, 60:0, 62:0, 67:0,
    38:0, 41:0, 46:0, 49:0, 58:0, 61:0, 70:0, 72:0,
    7:0, 29:0, 51:0,
    1:1, 3:1, 6:1, 16:1, 19:1, 22:1, 25:1, 26:1, 28:1,
    30:1, 31:1, 32:1, 33:1, 34:1, 39:1, 50:1, 53:1,
    55:1, 56:1, 63:1, 64:1, 68:1, 69:1, 71:1, 73:1, 74:1,
}

# ── Labels CSV ────────────────────────────────────────────
os.system('curl -L -o /kaggle/working/labels.zip "https://data.4tu.nl/file/b008dd74-020d-4ea4-a8ba-7bb60769d224/778d7d5f-6631-4cb6-9451-f88c574c7991"')
os.system('unzip -o /kaggle/working/labels.zip -d /kaggle/working/labels/')

col_names = ['recording', 'action_id', 'action_name', 'frame_start', 'frame_end']
train_df  = pd.read_csv('/kaggle/working/labels/train.csv', header=None, names=col_names)
val_df    = pd.read_csv('/kaggle/working/labels/val.csv',   header=None, names=col_names)
test_df   = pd.read_csv('/kaggle/working/labels/test.csv',  header=None, names=col_names)

for df in [train_df, val_df, test_df]:
    df['frame_start'] = df['frame_start'].str.replace('.jpg','').astype(int)
    df['frame_end']   = df['frame_end'].str.replace('.jpg','').astype(int)

# ── Charger pkl ───────────────────────────────────────────
with open(f'{PREP_DIR}scaler.pkl',            'rb') as f: scaler            = pickle.load(f)
with open(f'{PREP_DIR}stride_map_train.pkl',  'rb') as f: stride_map_train  = pickle.load(f)
with open(f'{PREP_DIR}stride_map_val.pkl',    'rb') as f: stride_map_val    = pickle.load(f)
with open(f'{PREP_DIR}rec_to_file_train.pkl', 'rb') as f: rec_to_file_train = pickle.load(f)
with open(f'{PREP_DIR}rec_to_file_val.pkl',   'rb') as f: rec_to_file_val   = pickle.load(f)

# ── Stride map test ───────────────────────────────────────
rec_to_file_test = {}
for f in os.listdir(TEST_FEAT_DIR):
    if f.endswith('.npy'):
        rec = re.sub(r'^test_p\d+_', '', f).replace('_features.npy', '')
        rec_to_file_test[rec] = os.path.join(TEST_FEAT_DIR, f)

stride_map_test = {}
for rec, path in rec_to_file_test.items():
    sub = test_df[test_df['recording'] == rec]
    if len(sub) == 0:
        continue
    arr    = np.load(path)
    f_min  = sub['frame_start'].min()
    f_max  = sub['frame_end'].max()
    n_feat = arr.shape[0]
    total  = f_max - f_min
    stride = round((total - 32) / (n_feat - 1)) if n_feat > 1 else 1
    stride = max(1, stride)
    stride_map_test[rec] = {'stride': stride, 'f_min': f_min, 'n_feats': n_feat}

print(f'Train:{len(stride_map_train)} Val:{len(stride_map_val)} Test:{len(stride_map_test)} ✅')

# ── PSR Error frames ──────────────────────────────────────
def load_psr_errors(psr_dir):
    files = glob.glob(f'{psr_dir}/**/*with_errors*', recursive=True)
    all_psr = []
    for f in files:
        rec = f.split('/')[-2]
        df  = pd.read_csv(f, header=None, names=['frame','step_id','description'])
        df['recording'] = rec
        all_psr.append(df)
    if not all_psr:
        return pd.DataFrame(), defaultdict(list)
    psr_df = pd.concat(all_psr, ignore_index=True)
    psr_df['is_error']  = psr_df['description'].str.contains(
        'Incorrect|Remove', case=False, na=False).astype(int)
    psr_df['frame_num'] = psr_df['frame'].str.replace('.jpg','').astype(int)
    err_dict = defaultdict(list)
    for _, row in psr_df[psr_df['is_error']==1].iterrows():
        err_dict[row['recording']].append(row['frame_num'])
    return psr_df, err_dict

psr_train_df, error_frames_train = load_psr_errors(PSR_TRAIN_DIR)
psr_val_df,   error_frames_val   = load_psr_errors(PSR_VAL_DIR)

print(f'Train errors : {psr_train_df["is_error"].sum() if len(psr_train_df) else 0}')
print(f'Val   errors : {psr_val_df["is_error"].sum() if len(psr_val_df) else 0}')

# ── Fonctions ─────────────────────────────────────────────
def get_feat_indices(frame_start, frame_end, f_min, n_feats, stride):
    i0 = max(0, round((frame_start - f_min) / stride))
    i1 = min(n_feats, round((frame_end   - f_min) / stride))
    return i0, i1

print('✅ Setup complet')
```

```python
# ── Charger modèle BiGRU ──────────────────────────────────
class TemporalAttention(nn.Module):
    def __init__(self, hidden_dim):
        super().__init__()
        self.attn = nn.Sequential(
            nn.Linear(hidden_dim * 2, hidden_dim),
            nn.Tanh(),
            nn.Linear(hidden_dim, 1)
        )
    def forward(self, gru_out, lengths=None):
        scores = self.attn(gru_out).squeeze(-1)
        if lengths is not None:
            for i, l in enumerate(lengths):
                if l < scores.shape[1]:
                    scores[i, l:] = float('-inf')
        return torch.softmax(scores, dim=1)

class ResidualBlock(nn.Module):
    def __init__(self, dim, dropout):
        super().__init__()
        self.block = nn.Sequential(
            nn.Linear(dim, dim), nn.LayerNorm(dim),
            nn.GELU(), nn.Dropout(dropout),
            nn.Linear(dim, dim), nn.LayerNorm(dim),
        )
        self.act = nn.GELU()
    def forward(self, x):
        return self.act(x + self.block(x))

class BiGRUDualHead(nn.Module):
    def __init__(self, input_dim, hidden_dim, num_layers, num_classes, dropout=0.4):
        super().__init__()
        self.input_proj = nn.Sequential(
            nn.Linear(input_dim, hidden_dim), nn.LayerNorm(hidden_dim),
            nn.GELU(), nn.Dropout(dropout), ResidualBlock(hidden_dim, dropout)
        )
        self.bigru = nn.GRU(
            input_size=hidden_dim, hidden_size=hidden_dim,
            num_layers=num_layers, batch_first=True,
            bidirectional=True,
            dropout=dropout if num_layers > 1 else 0.0
        )
        self.norm      = nn.LayerNorm(hidden_dim * 2)
        self.dropout   = nn.Dropout(dropout)
        self.attention = TemporalAttention(hidden_dim)
        self.action_proj     = nn.Sequential(
            nn.Linear(hidden_dim * 2, hidden_dim), nn.LayerNorm(hidden_dim),
            nn.GELU(), nn.Dropout(dropout)
        )
        self.action_residual = ResidualBlock(hidden_dim, dropout / 2)
        self.action_out      = nn.Linear(hidden_dim, num_classes)
        self.anomaly_head    = nn.Sequential(
            nn.Linear(hidden_dim * 2, hidden_dim // 2), nn.LayerNorm(hidden_dim // 2),
            nn.GELU(), nn.Dropout(dropout),
            nn.Linear(hidden_dim // 2, hidden_dim // 4),
            nn.GELU(), nn.Linear(hidden_dim // 4, 1), nn.Sigmoid()
        )

    def forward(self, x, lengths=None):
        B, T, _ = x.shape
        x = self.input_proj(x)
        if lengths is not None:
            packed = nn.utils.rnn.pack_padded_sequence(
                x, lengths.cpu(), batch_first=True, enforce_sorted=False)
            gru_out, _ = self.bigru(packed)
            gru_out, _ = nn.utils.rnn.pad_packed_sequence(
                gru_out, batch_first=True, total_length=T)
        else:
            gru_out, _ = self.bigru(x)
        gru_out      = self.norm(self.dropout(gru_out))
        attn_weights = self.attention(gru_out, lengths)
        action_feat  = self.action_residual(self.action_proj(gru_out))
        action_logits = self.action_out(action_feat)
        anomaly_score = self.anomaly_head(gru_out)
        return action_logits, anomaly_score, attn_weights, gru_out

# Charger checkpoint
ckpt   = torch.load(CKPT_PATH, map_location=DEVICE, weights_only=False)
model  = BiGRUDualHead(
    input_dim=FEAT_DIM, hidden_dim=512,
    num_layers=2, num_classes=NUM_CLASSES, dropout=0.4
).to(DEVICE)
model.load_state_dict(ckpt['model_state'])
model.eval()
print(f'✅ Modèle chargé | val_f1={ckpt["val_f1"]:.4f}')
```

```python
# ── Extraire hidden states + scores ───────────────────────
def extract_features_and_scores(df, rec_to_file, stride_map,
                                  scaler, model, device,
                                  error_frames_dict=None,
                                  clip_val=5.0, seq_len=256):
    """
    Retourne pour chaque frame :
    - gru_hidden : (N, 1024)
    - bigru_entropy : (N,) entropie des logits
    - is_error : (N,) 0=normal 1=erreur
    - recording : (N,) nom du recording
    """
    all_hidden, all_entropy, all_error, all_rec = [], [], [], []

    model.eval()
    with torch.no_grad():
        for rec, info in stride_map.items():
            sub = df[df['recording'] == rec]
            if len(sub) == 0:
                continue

            arr = np.load(rec_to_file[rec]).astype(np.float32)
            arr = scaler.transform(arr).astype(np.float32)
            arr = np.clip(arr, -clip_val, clip_val)
            n   = arr.shape[0]

            # Labels erreur frame-level
            error_mask = np.zeros(n, dtype=np.int64)
            if error_frames_dict and rec in error_frames_dict:
                for err_frame in error_frames_dict[rec]:
                    err_idx = round((err_frame - info['f_min']) / info['stride'])
                    err_idx = max(0, min(err_idx, n-1))
                    start   = max(0, err_idx - 15)
                    end     = min(n, err_idx + 15)
                    error_mask[start:end] = 1

            # Inférence par fenêtres
            hidden_seq  = np.zeros((n, 1024), dtype=np.float32)
            entropy_seq = np.zeros(n,         dtype=np.float32)
            counts      = np.zeros(n,         dtype=np.int32)

            for start in range(0, n - 32 + 1, 32):
                end   = min(start + seq_len, n)
                chunk = torch.tensor(arr[start:end]).unsqueeze(0).to(device)
                T     = chunk.shape[1]

                logits, _, _, hidden = model(chunk)
                probs   = F.softmax(logits, dim=-1).squeeze(0).cpu().numpy()
                entropy = -np.sum(probs * np.log(probs + 1e-8), axis=-1)
                h       = hidden.squeeze(0).cpu().numpy()

                hidden_seq[start:start+T]  += h
                entropy_seq[start:start+T] += entropy
                counts[start:start+T]      += 1

            # Moyenne
            valid = counts > 0
            hidden_seq[valid]  /= counts[valid, None]
            entropy_seq[valid] /= counts[valid]

            all_hidden.extend(hidden_seq[valid])
            all_entropy.extend(entropy_seq[valid])
            all_error.extend(error_mask[valid])
            all_rec.extend([rec] * valid.sum())

    return (np.array(all_hidden),
            np.array(all_entropy),
            np.array(all_error),
            np.array(all_rec))

print('Extraction train hidden states...')
train_hidden, train_entropy, train_error, train_recs = extract_features_and_scores(
    train_df, rec_to_file_train, stride_map_train,
    scaler, model, DEVICE,
    error_frames_dict=error_frames_train
)
print(f'Train : {train_hidden.shape} | erreurs : {train_error.sum()}')

print('\nExtraction val hidden states...')
val_hidden, val_entropy, val_error, val_recs = extract_features_and_scores(
    val_df, rec_to_file_val, stride_map_val,
    scaler, model, DEVICE,
    error_frames_dict=error_frames_val
)
print(f'Val : {val_hidden.shape} | erreurs : {val_error.sum()}')
```

```python
# ── Mahalanobis prototype bank ────────────────────────────
print('Construction prototypes Mahalanobis...')

# Utiliser seulement frames normales du train
normal_mask    = train_error == 0
X_train_normal = train_hidden[normal_mask]

print(f'Frames normales train : {X_train_normal.shape[0]}')

# Prototype global (comportement normal)
prototype  = X_train_normal.mean(axis=0)
var        = X_train_normal.var(axis=0) + 1e-4
inv_cov    = np.diag(1.0 / var)

# Score Mahalanobis sur val
print('Calcul scores Mahalanobis val...')
maha_scores_val = np.zeros(len(val_hidden))
for i, feat in enumerate(val_hidden):
    diff = feat - prototype
    maha_scores_val[i] = np.sqrt(np.maximum(diff @ inv_cov @ diff, 0))

print(f'Mahalanobis val | mean={maha_scores_val.mean():.3f} | std={maha_scores_val.std():.3f}')

# ── Score combiné ─────────────────────────────────────────
# Normaliser les scores
def normalize(x):
    return (x - x.min()) / (x.max() - x.min() + 1e-8)

maha_norm    = normalize(maha_scores_val)
entropy_norm = normalize(val_entropy)

# Combinaison pondérée
combined_score = 0.6 * maha_norm + 0.4 * entropy_norm

# ── AUC-ROC ───────────────────────────────────────────────
print('\n=== AUC-ROC RÉSULTATS ===')

if val_error.sum() > 0:
    auc_maha     = roc_auc_score(val_error, maha_scores_val)
    auc_entropy  = roc_auc_score(val_error, val_entropy)
    auc_combined = roc_auc_score(val_error, combined_score)

    print(f'AUC-ROC Mahalanobis  : {auc_maha:.4f}')
    print(f'AUC-ROC Entropy      : {auc_entropy:.4f}')
    print(f'AUC-ROC Combined     : {auc_combined:.4f}')

    # Courbe ROC
    plt.figure(figsize=(8, 6))
    for name, scores, auc in [
        ('Mahalanobis',  maha_scores_val, auc_maha),
        ('Entropy',      val_entropy,     auc_entropy),
        ('Combined',     combined_score,  auc_combined),
    ]:
        fpr, tpr, _ = roc_curve(val_error, scores)
        plt.plot(fpr, tpr, label=f'{name} (AUC={auc:.3f})')

    plt.plot([0,1], [0,1], 'k--', label='Random')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve — Assembly Error Detection')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()
    plt.savefig('/kaggle/working/roc_curve.png', dpi=100)
    plt.show()
    print(f'\n✅ Courbe ROC sauvegardée')
else:
    print('⚠️ Aucune erreur annotée dans val — vérifier PSR val labels')
```

```python
from sklearn.svm import OneClassSVM
from sklearn.neighbors import LocalOutlierFactor
from sklearn.preprocessing import StandardScaler as SS
from sklearn.decomposition import PCA, KernelPCA
from sklearn.ensemble import IsolationForest
from sklearn.metrics import roc_auc_score
import numpy as np

results = {}

# ── Préparation ───────────────────────────────────────────
X_train_n = train_hidden[train_error == 0]  # normales seulement
X_val     = val_hidden

# Standardiser
ss = SS()
X_train_ss = ss.fit_transform(X_train_n)
X_val_ss   = ss.transform(X_val)

print(f'Train normal : {X_train_ss.shape}')
print(f'Val          : {X_val_ss.shape}')
print(f'Val erreurs  : {val_error.sum()}')

# ── Test différentes dimensions PCA ───────────────────────
print('\n=== PCA DIMENSIONS ===')
for n_comp in [32, 64, 128, 256]:
    pca  = PCA(n_components=n_comp, random_state=42)
    Xtr  = pca.fit_transform(X_train_ss)
    Xval = pca.transform(X_val_ss)
    var  = Xtr.var(axis=0) + 1e-4
    proto = Xtr.mean(axis=0)
    inv  = np.diag(1.0 / var)
    scores = np.array([
        np.sqrt(np.maximum((v-proto) @ inv @ (v-proto), 0))
        for v in Xval
    ])
    auc = roc_auc_score(val_error, scores)
    results[f'Mahalanobis_PCA{n_comp}'] = auc
    print(f'  PCA{n_comp:>3} Mahalanobis : {auc:.4f}')

# ── Isolation Forest différents params ────────────────────
print('\n=== ISOLATION FOREST ===')
pca64  = PCA(n_components=64, random_state=42)
Xtr64  = pca64.fit_transform(X_train_ss)
Xval64 = pca64.transform(X_val_ss)

for n_est, contam in [(100,0.05),(200,0.05),(200,0.10),(500,0.05)]:
    iso = IsolationForest(n_estimators=n_est, contamination=contam,
                          random_state=42, n_jobs=-1)
    iso.fit(Xtr64)
    scores = -iso.score_samples(Xval64)
    auc = roc_auc_score(val_error, scores)
    results[f'IsoForest_n{n_est}_c{contam}'] = auc
    print(f'  n_est={n_est} contam={contam} : {auc:.4f}')

# ── One-Class SVM ─────────────────────────────────────────
print('\n=== ONE-CLASS SVM ===')
for nu, kernel in [(0.05,'rbf'),(0.10,'rbf'),(0.05,'linear')]:
    try:
        ocsvm = OneClassSVM(nu=nu, kernel=kernel)
        ocsvm.fit(Xtr64[:5000])  # subsample pour vitesse
        scores = -ocsvm.score_samples(Xval64)
        auc = roc_auc_score(val_error, scores)
        results[f'OCSVM_{kernel}_nu{nu}'] = auc
        print(f'  {kernel} nu={nu} : {auc:.4f}')
    except Exception as e:
        print(f'  {kernel} nu={nu} : erreur {e}')

# ── Local Outlier Factor ───────────────────────────────────
print('\n=== LOCAL OUTLIER FACTOR ===')
for n_neighbors in [5, 10, 20]:
    try:
        lof = LocalOutlierFactor(n_neighbors=n_neighbors,
                                  novelty=True, n_jobs=-1)
        lof.fit(Xtr64)
        scores = -lof.score_samples(Xval64)
        auc = roc_auc_score(val_error, scores)
        results[f'LOF_k{n_neighbors}'] = auc
        print(f'  k={n_neighbors} : {auc:.4f}')
    except Exception as e:
        print(f'  k={n_neighbors} : erreur {e}')

# ── Mahalanobis par classe ────────────────────────────────
print('\n=== MAHALANOBIS PAR CLASSE ───')
# Extraire classe prédite par BiGRU sur val
val_preds = []
model.eval()
with torch.no_grad():
    for rec, info in stride_map_val.items():
        arr = np.load(rec_to_file_val[rec]).astype(np.float32)
        arr = scaler.transform(arr)
        arr = np.clip(arr, -5.0, 5.0)
        n   = arr.shape[0]
        preds_rec = np.zeros(n, dtype=np.int64)
        for start in range(0, n-32+1, 32):
            end   = min(start+256, n)
            chunk = torch.tensor(arr[start:end]).unsqueeze(0)
            logits, _, _, _ = model(chunk)
            p = logits.argmax(dim=-1).squeeze(0).numpy()
            preds_rec[start:start+len(p)] = p
        valid = np.ones(n, dtype=bool)
        val_preds.extend(preds_rec[valid])

val_preds = np.array(val_preds[:len(val_hidden)])

# Prototypes par classe
pca128 = PCA(n_components=128, random_state=42)
Xtr128 = pca128.fit_transform(X_train_ss)
Xval128 = pca128.transform(X_val_ss)

train_preds_all = []
model.eval()
with torch.no_grad():
    for rec, info in stride_map_train.items():
        arr = np.load(rec_to_file_train[rec]).astype(np.float32)
        arr = scaler.transform(arr)
        arr = np.clip(arr, -5.0, 5.0)
        n   = arr.shape[0]
        preds_rec = np.zeros(n, dtype=np.int64)
        for start in range(0, n-32+1, 32):
            end   = min(start+256, n)
            chunk = torch.tensor(arr[start:end]).unsqueeze(0)
            logits, _, _, _ = model(chunk)
            p = logits.argmax(dim=-1).squeeze(0).numpy()
            preds_rec[start:start+len(p)] = p
        train_preds_all.extend(preds_rec)

train_preds_all = np.array(train_preds_all[:len(train_hidden)])
train_normal_mask = train_error == 0

scores_per_class = np.zeros(len(val_hidden))
for cls in range(NUM_CLASSES):
    train_cls_mask = (train_preds_all == cls) & train_normal_mask
    if train_cls_mask.sum() < 10:
        continue
    X_cls   = Xtr128[train_cls_mask[:len(Xtr128)]]
    proto   = X_cls.mean(axis=0)
    var     = X_cls.var(axis=0) + 1e-4
    inv     = np.diag(1.0 / var)
    val_cls = val_preds == cls
    for i in np.where(val_cls[:len(Xval128)])[0]:
        diff = Xval128[i] - proto
        scores_per_class[i] = np.sqrt(np.maximum(diff @ inv @ diff, 0))

auc_per_class = roc_auc_score(val_error, scores_per_class)
results['Mahalanobis_per_class'] = auc_per_class
print(f'  Mahalanobis par classe : {auc_per_class:.4f}')

# ── Résumé ────────────────────────────────────────────────
print(f'\n=== RÉSUMÉ TOUTES MÉTHODES ===')
for name, auc in sorted(results.items(), key=lambda x: -x[1]):
    flag = '✅' if auc > 0.65 else ('⚠️' if auc > 0.60 else '❌')
    print(f'  {flag} {name:40s} : {auc:.4f}')

best_name = max(results, key=results.get)
print(f'\nMeilleure méthode : {best_name} → AUC={results[best_name]:.4f}')
```

```python
# ── Viterbi log-likelihood score ─────────────────────────
print('=== VITERBI LOG-LIKELIHOOD ===')

# Construire matrice transitions depuis train normal
transition_counts = np.zeros((NUM_CLASSES, NUM_CLASSES))

for rec, info in stride_map_train.items():
    sub = train_df[train_df['recording'] == rec]
    if len(sub) == 0:
        continue
    arr    = np.load(rec_to_file_train[rec]).astype(np.float32)
    arr    = scaler.transform(arr)
    arr    = np.clip(arr, -5.0, 5.0)
    n      = arr.shape[0]

    labels = np.full(n, -1, dtype=np.int64)
    for _, row in sub.iterrows():
        if row['action_id'] not in SEMANTIC_GROUPS_2:
            continue
        i0, i1 = get_feat_indices(
            row['frame_start'], row['frame_end'],
            info['f_min'], info['n_feats'], info['stride']
        )
        labels[i0:i1] = SEMANTIC_GROUPS_2[row['action_id']]

    # Compter transitions
    prev = -1
    for lbl in labels:
        if lbl >= 0:
            if prev >= 0:
                transition_counts[prev, lbl] += 1
            prev = lbl

# Normaliser
transition_counts += 0.1  # Laplace smoothing
row_sums = transition_counts.sum(axis=1, keepdims=True)
trans_matrix = transition_counts / row_sums
log_trans    = np.log(trans_matrix)

init_counts = transition_counts.sum(axis=0)
log_init    = np.log(init_counts / init_counts.sum())

print(f'Transition matrix :\n{trans_matrix.round(3)}')

# ── Score Viterbi : log-likelihood par frame ──────────────
def viterbi_frame_scores(log_probs_seq, log_trans, log_init):
    """
    Retourne un score d'anomalie par frame
    = probabilité négative dans le chemin Viterbi
    """
    T, C = log_probs_seq.shape
    viterbi = np.zeros((T, C))
    backptr = np.zeros((T, C), dtype=int)

    viterbi[0] = log_init + log_probs_seq[0]
    for t in range(1, T):
        for c in range(C):
            scores = viterbi[t-1] + log_trans[:, c]
            viterbi[t, c]  = scores.max() + log_probs_seq[t, c]
            backptr[t, c]  = scores.argmax()

    # Backtrack
    path = np.zeros(T, dtype=int)
    path[-1] = viterbi[-1].argmax()
    for t in range(T-2, -1, -1):
        path[t] = backptr[t+1, path[t+1]]

    # Score anomalie = -log_prob du chemin optimal
    path_scores = np.array([
        -viterbi[t, path[t]] for t in range(T)
    ])
    return path_scores, path

# Calculer scores Viterbi sur val
print('\nCalcul Viterbi scores val...')
viterbi_scores_all = []

model.eval()
with torch.no_grad():
    for rec, info in stride_map_val.items():
        arr = np.load(rec_to_file_val[rec]).astype(np.float32)
        arr = scaler.transform(arr)
        arr = np.clip(arr, -5.0, 5.0)
        n   = arr.shape[0]

        # Inférence complète
        all_logprobs = np.zeros((n, NUM_CLASSES))
        counts       = np.zeros(n)

        for start in range(0, n-32+1, 32):
            end   = min(start+256, n)
            chunk = torch.tensor(arr[start:end]).unsqueeze(0)
            logits, _, _, _ = model(chunk)
            lp = torch.log_softmax(logits, dim=-1).squeeze(0).numpy()
            T  = lp.shape[0]
            all_logprobs[start:start+T] += lp
            counts[start:start+T]       += 1

        valid = counts > 0
        all_logprobs[valid] /= counts[valid, None]

        # Viterbi scores
        lp_valid = all_logprobs[valid]
        v_scores, _ = viterbi_frame_scores(lp_valid, log_trans, log_init)
        viterbi_scores_all.extend(v_scores)

viterbi_scores_all = np.array(viterbi_scores_all[:len(val_error)])
auc_viterbi = roc_auc_score(val_error, viterbi_scores_all)
print(f'AUC-ROC Viterbi log-likelihood : {auc_viterbi:.4f}')

# ── Combinaison Mahalanobis + Viterbi ────────────────────
print('\n=== COMBINAISON MAHALANOBIS + VITERBI ===')

# Meilleur Mahalanobis : PCA64
pca64_final = PCA(n_components=64, random_state=42)
Xtr_f = pca64_final.fit_transform(X_train_ss)
Xval_f = pca64_final.transform(X_val_ss)
proto_f = Xtr_f.mean(axis=0)
var_f   = Xtr_f.var(axis=0) + 1e-4
inv_f   = np.diag(1.0 / var_f)
maha_final = np.array([
    np.sqrt(np.maximum((v-proto_f) @ inv_f @ (v-proto_f), 0))
    for v in Xval_f
])

def norm(x): return (x - x.min()) / (x.max() - x.min() + 1e-8)

maha_n   = norm(maha_final)
viterbi_n = norm(viterbi_scores_all)

for w in [0.5, 0.6, 0.7, 0.8, 0.9]:
    combo = w * maha_n + (1-w) * viterbi_n
    auc   = roc_auc_score(val_error, combo)
    print(f'  Maha={w:.1f} Viterbi={1-w:.1f} : {auc:.4f}')

# Résumé
print(f'\n=== RÉSUMÉ FINAL ===')
print(f'Mahalanobis PCA64   : 0.6684')
print(f'Viterbi likelihood  : {auc_viterbi:.4f}')
best_combo = max(
    [(w, roc_auc_score(val_error, w*maha_n + (1-w)*viterbi_n))
     for w in [0.5,0.6,0.7,0.8,0.9]],
    key=lambda x: x[1]
)
print(f'Meilleure combinaison Maha+Viterbi : {best_combo[1]:.4f} (w={best_combo[0]})')
```

```python
# ── Viterbi log-likelihood score ─────────────────────────
print('=== VITERBI LOG-LIKELIHOOD ===')

# Construire matrice transitions depuis train normal
transition_counts = np.zeros((NUM_CLASSES, NUM_CLASSES))

for rec, info in stride_map_train.items():
    sub = train_df[train_df['recording'] == rec]
    if len(sub) == 0:
        continue
    arr    = np.load(rec_to_file_train[rec]).astype(np.float32)
    arr    = scaler.transform(arr)
    arr    = np.clip(arr, -5.0, 5.0)
    n      = arr.shape[0]

    labels = np.full(n, -1, dtype=np.int64)
    for _, row in sub.iterrows():
        if row['action_id'] not in SEMANTIC_GROUPS_2:
            continue
        i0, i1 = get_feat_indices(
            row['frame_start'], row['frame_end'],
            info['f_min'], info['n_feats'], info['stride']
        )
        labels[i0:i1] = SEMANTIC_GROUPS_2[row['action_id']]

    prev = -1
    for lbl in labels:
        if lbl >= 0:
            if prev >= 0:
                transition_counts[prev, lbl] += 1
            prev = lbl

transition_counts += 0.1
row_sums     = transition_counts.sum(axis=1, keepdims=True)
trans_matrix = transition_counts / row_sums
log_trans    = np.log(trans_matrix)
init_counts  = transition_counts.sum(axis=0)
log_init     = np.log(init_counts / init_counts.sum())

print(f'Transition matrix :\n{trans_matrix.round(3)}')

# ── Viterbi frame scores ──────────────────────────────────
def viterbi_frame_scores(log_probs_seq, log_trans, log_init):
    T, C = log_probs_seq.shape
    viterbi = np.zeros((T, C))
    backptr = np.zeros((T, C), dtype=int)

    viterbi[0] = log_init + log_probs_seq[0]
    for t in range(1, T):
        for c in range(C):
            scores        = viterbi[t-1] + log_trans[:, c]
            viterbi[t, c] = scores.max() + log_probs_seq[t, c]
            backptr[t, c] = scores.argmax()

    path = np.zeros(T, dtype=int)
    path[-1] = viterbi[-1].argmax()
    for t in range(T-2, -1, -1):
        path[t] = backptr[t+1, path[t+1]]

    # Score anomalie = log-prob NÉGATIF du chemin optimal
    # → frame improbable = score élevé = anomalie
    path_scores = np.array([
        -viterbi[t, path[t]] for t in range(T)
    ])
    return path_scores, path

# ── Calcul scores Viterbi sur val ─────────────────────────
print('\nCalcul Viterbi scores val...')
viterbi_scores_all = []

model.eval()
with torch.no_grad():
    for rec, info in stride_map_val.items():
        arr = np.load(rec_to_file_val[rec]).astype(np.float32)
        arr = scaler.transform(arr)
        arr = np.clip(arr, -5.0, 5.0)
        n   = arr.shape[0]

        all_logprobs = np.zeros((n, NUM_CLASSES))
        counts       = np.zeros(n)

        for start in range(0, n-32+1, 32):
            end   = min(start+256, n)
            chunk = torch.tensor(arr[start:end]).unsqueeze(0)
            logits, _, _, _ = model(chunk)
            lp = torch.log_softmax(logits, dim=-1).squeeze(0).numpy()
            T  = lp.shape[0]
            all_logprobs[start:start+T] += lp
            counts[start:start+T]       += 1

        valid = counts > 0
        all_logprobs[valid] /= counts[valid, None]

        lp_valid = all_logprobs[valid]
        v_scores, _ = viterbi_frame_scores(lp_valid, log_trans, log_init)
        viterbi_scores_all.extend(v_scores)

viterbi_scores_all = np.array(viterbi_scores_all[:len(val_error)])

# AUC direct et inversé
auc_viterbi     = roc_auc_score(val_error, viterbi_scores_all)
auc_viterbi_inv = roc_auc_score(val_error, -viterbi_scores_all)
print(f'AUC-ROC Viterbi          : {auc_viterbi:.4f}')
print(f'AUC-ROC Viterbi inversé  : {auc_viterbi_inv:.4f}')

# Utiliser le meilleur sens
best_viterbi_scores = viterbi_scores_all if auc_viterbi > auc_viterbi_inv \
                      else -viterbi_scores_all
best_viterbi_auc    = max(auc_viterbi, auc_viterbi_inv)
print(f'AUC-ROC Viterbi meilleur : {best_viterbi_auc:.4f}')

# ── Mahalanobis PCA64 (meilleur) ──────────────────────────
pca64_final = PCA(n_components=64, random_state=42)
Xtr_f  = pca64_final.fit_transform(X_train_ss)
Xval_f = pca64_final.transform(X_val_ss)
proto_f = Xtr_f.mean(axis=0)
var_f   = Xtr_f.var(axis=0) + 1e-4
inv_f   = np.diag(1.0 / var_f)
maha_final = np.array([
    np.sqrt(np.maximum((v-proto_f) @ inv_f @ (v-proto_f), 0))
    for v in Xval_f
])
auc_maha = roc_auc_score(val_error, maha_final)
print(f'\nAUC-ROC Mahalanobis PCA64 : {auc_maha:.4f}')

# ── Combinaison Mahalanobis + Viterbi ────────────────────
def norm(x): return (x - x.min()) / (x.max() - x.min() + 1e-8)

maha_n    = norm(maha_final)
viterbi_n = norm(best_viterbi_scores)

print('\n=== COMBINAISON MAHALANOBIS + VITERBI ===')
best_combo_auc  = 0
best_combo_w    = 0
best_combo_scores = None

for w in [0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9]:
    combo = w * maha_n + (1-w) * viterbi_n
    auc   = roc_auc_score(val_error, combo)
    flag  = '✅' if auc > 0.68 else ''
    print(f'  Maha={w:.1f} Viterbi={1-w:.1f} : {auc:.4f} {flag}')
    if auc > best_combo_auc:
        best_combo_auc    = auc
        best_combo_w      = w
        best_combo_scores = combo

# ── Résumé final + courbe ROC ─────────────────────────────
print(f'\n=== RÉSUMÉ FINAL ===')
print(f'Mahalanobis PCA64        : {auc_maha:.4f}')
print(f'Viterbi meilleur         : {best_viterbi_auc:.4f}')
print(f'Combinaison optimale     : {best_combo_auc:.4f} (w_maha={best_combo_w})')

import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve

plt.figure(figsize=(8, 6))
for name, scores, auc in [
    ('Mahalanobis PCA64', maha_final,           auc_maha),
    ('Viterbi',           best_viterbi_scores,   best_viterbi_auc),
    ('Combined',          best_combo_scores,     best_combo_auc),
]:
    fpr, tpr, _ = roc_curve(val_error, scores)
    plt.plot(fpr, tpr, label=f'{name} (AUC={auc:.3f})', linewidth=2)

plt.plot([0,1], [0,1], 'k--', label='Random (AUC=0.500)', linewidth=1)
plt.xlabel('False Positive Rate', fontsize=12)
plt.ylabel('True Positive Rate',  fontsize=12)
plt.title('ROC Curve — Assembly Error Detection\nIndustReal Dataset', fontsize=13)
plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.savefig('/kaggle/working/roc_curve_final.png', dpi=150)
plt.show()
print('✅ Courbe ROC sauvegardée')
```

Les erreurs en début de séquence → bien détectées ✅
Les erreurs en fin de séquence   → mal détectées ❌

Cause Viterbi drift :
→ Viterbi log-likelihood cumulatif
→ score diminue naturellement au fil du temps
→ biais temporel non corrigé

Fix possible :
→ Viterbi par fenêtre glissante
→ au lieu de séquence entière

```python
# ── Viterbi fenêtre glissante ─────────────────────────────
def viterbi_sliding_window(log_probs_seq, log_trans, log_init, 
                            window=50):
    """
    Score anomalie par fenêtre glissante
    → pas de drift temporel
    → compatible temps réel
    """
    T, C     = log_probs_seq.shape
    scores   = np.zeros(T)

    for t in range(T):
        # Fenêtre [t-window, t]
        start = max(0, t - window)
        chunk = log_probs_seq[start:t+1]
        W, _  = chunk.shape

        if W < 2:
            scores[t] = 0
            continue

        # Viterbi sur cette fenêtre
        v = np.zeros((W, C))
        v[0] = log_init + chunk[0]
        for i in range(1, W):
            for c in range(C):
                sc     = v[i-1] + log_trans[:, c]
                v[i,c] = sc.max() + chunk[i, c]

        # Score = -log-likelihood de la dernière frame
        scores[t] = -v[-1].max()

    return scores

# ── Tester différentes tailles de fenêtre ────────────────
print('=== VITERBI FENÊTRE GLISSANTE ===')

viterbi_sliding_results = {}

model.eval()
with torch.no_grad():
    for window_size in [20, 30, 50, 100]:
        all_scores = []

        for rec, info in stride_map_val.items():
            arr = np.load(rec_to_file_val[rec]).astype(np.float32)
            arr = scaler.transform(arr)
            arr = np.clip(arr, -5.0, 5.0)
            n   = arr.shape[0]

            all_logprobs = np.zeros((n, NUM_CLASSES))
            counts       = np.zeros(n)

            for start in range(0, n-32+1, 32):
                end   = min(start+256, n)
                chunk = torch.tensor(arr[start:end]).unsqueeze(0)
                logits, _, _, _ = model(chunk)
                lp = torch.log_softmax(logits, dim=-1).squeeze(0).numpy()
                T  = lp.shape[0]
                all_logprobs[start:start+T] += lp
                counts[start:start+T]       += 1

            valid = counts > 0
            all_logprobs[valid] /= counts[valid, None]

            # Viterbi glissant
            v_scores = viterbi_sliding_window(
                all_logprobs[valid], log_trans, log_init,
                window=window_size
            )
            all_scores.extend(v_scores)

        all_scores = np.array(all_scores[:len(val_error)])
        auc = roc_auc_score(val_error, all_scores)
        viterbi_sliding_results[window_size] = (auc, all_scores)
        print(f'  Window={window_size:>4} : AUC={auc:.4f}')

# Meilleure fenêtre
best_window = max(viterbi_sliding_results, 
                   key=lambda k: viterbi_sliding_results[k][0])
best_auc_sliding, best_scores_sliding = viterbi_sliding_results[best_window]
print(f'\nMeilleure fenêtre : {best_window} → AUC={best_auc_sliding:.4f}')
print(f'Viterbi complet   : 0.7506')
print(f'Gain              : {best_auc_sliding - 0.7506:+.4f}')

# ── Nouvelle combinaison avec fenêtre glissante ───────────
print('\n=== COMBINAISON AVEC VITERBI GLISSANT ===')
def norm(x): return (x - x.min()) / (x.max() - x.min() + 1e-8)

maha_n        = norm(maha_final)
viterbi_sl_n  = norm(best_scores_sliding)

best_combo_sliding = 0
best_w_sliding     = 0
best_scores_combo  = None

for w in [0.3, 0.4, 0.5, 0.6, 0.7]:
    combo = w * maha_n + (1-w) * viterbi_sl_n
    auc   = roc_auc_score(val_error, combo)
    flag  = '✅' if auc > 0.77 else ''
    print(f'  Maha={w:.1f} VitSlide={1-w:.1f} : {auc:.4f} {flag}')
    if auc > best_combo_sliding:
        best_combo_sliding = auc
        best_w_sliding     = w
        best_scores_combo  = combo

print(f'\nMeilleur combiné glissant : {best_combo_sliding:.4f}')
print(f'Ancien combiné            : 0.7685')
print(f'Gain                      : {best_combo_sliding - 0.7685:+.4f}')
```

```python
# ── Viterbi sans drift : score par frame indépendant ──────
def viterbi_local_score(log_probs_seq, log_trans, log_init, window=50):
    """
    Score local = log-likelihood de la frame t
    dans le contexte de ses W voisins SEULEMENT
    Reset à chaque fenêtre → pas de drift
    """
    T, C   = log_probs_seq.shape
    scores = np.zeros(T)

    for t in range(T):
        start = max(0, t - window // 2)
        end   = min(T, t + window // 2 + 1)
        chunk = log_probs_seq[start:end]
        W     = len(chunk)
        t_local = t - start  # position locale dans la fenêtre

        if W < 2:
            scores[t] = 0
            continue

        # Viterbi sur fenêtre centrée sur t
        v = np.zeros((W, C))
        v[0] = log_init + chunk[0]
        for i in range(1, W):
            for c in range(C):
                sc     = v[i-1] + log_trans[:, c]
                v[i,c] = sc.max() + chunk[i, c]

        # Score = -log-prob de la frame centrale uniquement
        scores[t] = -v[t_local].max()

    return scores

# ── Test fenêtre centrée ──────────────────────────────────
print('=== VITERBI FENÊTRE CENTRÉE (sans drift) ===')

model.eval()
centered_results = {}

with torch.no_grad():
    for window_size in [20, 30, 50, 100, 150]:
        scores_w = []
        for rec, info in stride_map_val.items():
            lp, valid = all_logprobs_val[rec]
            lp_v = lp[valid]
            v_sc = viterbi_local_score(
                lp_v, log_trans, log_init, window=window_size)
            scores_w.extend(v_sc)

        scores_w = np.array(scores_w[:len(val_error)])
        auc      = roc_auc_score(val_error, scores_w)
        auc_inv  = roc_auc_score(val_error, -scores_w)
        best_auc = max(auc, auc_inv)
        best_sc  = scores_w if auc > auc_inv else -scores_w
        centered_results[window_size] = (best_auc, best_sc)
        print(f'  Window={window_size:>4} : AUC={best_auc:.4f}')

# Meilleure fenêtre centrée
best_w   = max(centered_results, key=lambda k: centered_results[k][0])
best_auc_centered, best_sc_centered = centered_results[best_w]
print(f'\nMeilleure fenêtre centrée : {best_w} → AUC={best_auc_centered:.4f}')
print(f'Viterbi complet           : 0.7506')

# Combinaison avec Mahalanobis
print('\n=== COMBINAISON MAHA + VITERBI CENTRÉ ===')
vit_centered_n = norm(best_sc_centered)

best_final = 0
best_final_scores = None
for w in [0.3, 0.4, 0.5, 0.6, 0.7]:
    combo = w * maha_n + (1-w) * vit_centered_n
    auc   = roc_auc_score(val_error, combo)
    flag  = '✅' if auc > 0.77 else ''
    print(f'  Maha={w:.1f} VitCentered={1-w:.1f} : {auc:.4f} {flag}')
    if auc > best_final:
        best_final        = auc
        best_final_scores = combo

print(f'\nMeilleur final  : {best_final:.4f}')
print(f'Référence       : 0.7685')
print(f'Gain            : {best_final - 0.7685:+.4f}')
```

```python
# Normaliser Viterbi par position dans la séquence
# pour supprimer le biais temporel

def viterbi_detrended(log_probs_seq, log_trans, log_init):
    T, C = log_probs_seq.shape
    
    # Score Viterbi complet
    v_scores, path = viterbi_frame_scores(
        log_probs_seq, log_trans, log_init)
    v_inv = -v_scores
    
    # Détrending : soustraire tendance linéaire
    x    = np.arange(T)
    coef = np.polyfit(x, v_inv, 1)  # fit linéaire
    trend = np.polyval(coef, x)
    detrended = v_inv - trend
    
    return detrended

print('=== VITERBI DÉTRENDÉ ===')
detrended_scores = []

for rec, info in stride_map_val.items():
    lp, valid = all_logprobs_val[rec]
    lp_v = lp[valid]
    d_sc = viterbi_detrended(lp_v, log_trans, log_init)
    detrended_scores.extend(d_sc)

detrended_scores = np.array(detrended_scores[:len(val_error)])
auc_det     = roc_auc_score(val_error,  detrended_scores)
auc_det_inv = roc_auc_score(val_error, -detrended_scores)
best_auc_det = max(auc_det, auc_det_inv)
best_det_sc  = detrended_scores if auc_det > auc_det_inv \
               else -detrended_scores

print(f'AUC Viterbi détrendé : {best_auc_det:.4f}')
print(f'AUC Mahalanobis      : 0.6684')

# Combinaison
vit_det_n = norm(best_det_sc)
for w in [0.4, 0.5, 0.6]:
    combo = w * maha_n + (1-w) * vit_det_n
    auc   = roc_auc_score(val_error, combo)
    print(f'Maha={w} + VitDétrend={1-w} : {auc:.4f}')
```

```python
import os, glob
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler as SS
from sklearn.decomposition import PCA
from sklearn.metrics import roc_auc_score
from collections import defaultdict

# ── Charger gaze + hands et aligner avec stride_map ──────
def load_multimodal_features(rec, stride_map, modal_dir, 
                              error_frames_dict=None, clip_val=5.0):
    """
    Charge gaze.csv + hands.csv pour un recording
    Aligne avec stride_map → features par feature-frame
    """
    info = stride_map[rec]
    n    = info['n_feats']
    
    # Chercher les fichiers
    gaze_path  = None
    hands_path = None
    for root, dirs, files in os.walk(modal_dir):
        if os.path.basename(root) == rec:
            for f in files:
                if f == 'gaze.csv':
                    gaze_path = os.path.join(root, f)
                if f == 'hands.csv':
                    hands_path = os.path.join(root, f)

    gaze_feat  = np.zeros((n, 2),  dtype=np.float32)
    hands_feat = np.zeros((n, 42), dtype=np.float32)

    # ── Gaze ─────────────────────────────────────────────
    if gaze_path:
        df_g = pd.read_csv(gaze_path, header=None)
        df_g.columns = ['frame', 'x', 'y']
        df_g['frame_num'] = df_g['frame'].str.replace('.jpg','').astype(int)

        for _, row in df_g.iterrows():
            feat_idx = round((row['frame_num'] - info['f_min']) / info['stride'])
            feat_idx = max(0, min(feat_idx, n-1))
            gaze_feat[feat_idx, 0] = row['x']
            gaze_feat[feat_idx, 1] = row['y']

    # ── Hands ─────────────────────────────────────────────
    if hands_path:
        df_h = pd.read_csv(hands_path, header=None)
        df_h['frame_num'] = df_h.iloc[:, 0].str.replace('.jpg','').astype(int)
        # 21 joints × 2 coords = 42 features (ignorer z)
        joint_cols = []
        for j in range(21):
            col_x = 1 + j * 2
            col_y = 2 + j * 2
            if col_x < df_h.shape[1] and col_y < df_h.shape[1]:
                joint_cols.extend([col_x, col_y])

        for _, row in df_h.iterrows():
            feat_idx = round((row['frame_num'] - info['f_min']) / info['stride'])
            feat_idx = max(0, min(feat_idx, n-1))
            for k, col in enumerate(joint_cols[:42]):
                hands_feat[feat_idx, k] = row.iloc[col]

    # ── Error mask ────────────────────────────────────────
    err_mask = np.zeros(n, dtype=bool)
    if error_frames_dict and rec in error_frames_dict:
        for ef in error_frames_dict[rec]:
            ei = round((ef - info['f_min']) / info['stride'])
            ei = max(0, min(ei, n-1))
            err_mask[max(0,ei-15):min(n,ei+15)] = True

    return gaze_feat, hands_feat, err_mask


# ── Construire features multimodales train ────────────────
print('Construction features multimodales train...')

MODAL_TRAIN_DIR = '/kaggle/working/multimodal_train/'
MODAL_VAL_DIR   = '/kaggle/working/multimodal_val/'

X_gaze_train,  X_hands_train  = [], []
X_gaze_val,    X_hands_val    = [], []

# Train + Test (merged comme avant)
all_train_stride = {**stride_map_train, **stride_map_test}

for rec, info in all_train_stride.items():
    gaze_f, hands_f, err_mask = load_multimodal_features(
        rec, all_train_stride, MODAL_TRAIN_DIR,
        error_frames_dict=error_frames_train
    )
    # Garder seulement frames normales
    X_gaze_train.append(gaze_f[~err_mask])
    X_hands_train.append(hands_f[~err_mask])

X_gaze_train  = np.vstack(X_gaze_train)
X_hands_train = np.vstack(X_hands_train)

# Val
for rec, info in stride_map_val.items():
    gaze_f, hands_f, _ = load_multimodal_features(
        rec, stride_map_val, MODAL_VAL_DIR
    )
    X_gaze_val.append(gaze_f)
    X_hands_val.append(hands_f)

X_gaze_val  = np.vstack(X_gaze_val)
X_hands_val = np.vstack(X_hands_val)

print(f'Gaze  train : {X_gaze_train.shape}')
print(f'Hands train : {X_hands_train.shape}')
print(f'Gaze  val   : {X_gaze_val.shape}')
print(f'Hands val   : {X_hands_val.shape}')

# ── Normaliser ────────────────────────────────────────────
ss_gaze  = SS()
ss_hands = SS()

X_gaze_train_n  = ss_gaze.fit_transform(X_gaze_train)
X_hands_train_n = ss_hands.fit_transform(X_hands_train)
X_gaze_val_n    = ss_gaze.transform(X_gaze_val)
X_hands_val_n   = ss_hands.transform(X_hands_val)

# ── Mahalanobis sur gaze seul ─────────────────────────────
proto_gaze = X_gaze_train_n.mean(axis=0)
var_gaze   = X_gaze_train_n.var(axis=0) + 1e-4
inv_gaze   = np.diag(1.0 / var_gaze)

maha_gaze = np.array([
    np.sqrt(np.maximum((v-proto_gaze) @ inv_gaze @ (v-proto_gaze), 0))
    for v in X_gaze_val_n
])[:len(val_error)]
auc_gaze = roc_auc_score(val_error, maha_gaze)
print(f'\nAUC Gaze seul         : {auc_gaze:.4f}')

# ── Mahalanobis sur hands seul ────────────────────────────
pca_hands = PCA(n_components=20, random_state=42)
Xh_tr = pca_hands.fit_transform(X_hands_train_n)
Xh_vl = pca_hands.transform(X_hands_val_n)

proto_hands = Xh_tr.mean(axis=0)
var_hands   = Xh_tr.var(axis=0) + 1e-4
inv_hands   = np.diag(1.0 / var_hands)

maha_hands = np.array([
    np.sqrt(np.maximum((v-proto_hands) @ inv_hands @ (v-proto_hands), 0))
    for v in Xh_vl
])[:len(val_error)]
auc_hands = roc_auc_score(val_error, maha_hands)
print(f'AUC Hands seul        : {auc_hands:.4f}')

# ── Combinaison RGB + Gaze + Hands ────────────────────────
print('\n=== COMBINAISON RGB + GAZE + HANDS ===')

def norm(x): return (x - x.min()) / (x.max() - x.min() + 1e-8)

maha_rgb_n   = norm(maha_final)       # GRU PCA64 = 0.668
maha_gaze_n  = norm(maha_gaze)
maha_hands_n = norm(maha_hands)

best_auc_multi = 0
best_combo     = None

for w_rgb in [0.5, 0.6, 0.7]:
    for w_gaze in [0.1, 0.2, 0.3]:
        w_hands = 1 - w_rgb - w_gaze
        if w_hands < 0:
            continue
        combo = w_rgb * maha_rgb_n + w_gaze * maha_gaze_n + w_hands * maha_hands_n
        auc   = roc_auc_score(val_error, combo)
        if auc > best_auc_multi:
            best_auc_multi = auc
            best_combo     = combo
            best_w         = (w_rgb, w_gaze, w_hands)
        print(f'  RGB={w_rgb} Gaze={w_gaze} Hands={w_hands:.1f} : {auc:.4f}')

print(f'\nMeilleur multimodal : {best_auc_multi:.4f}')
print(f'Weights             : RGB={best_w[0]} Gaze={best_w[1]} Hands={best_w[2]:.1f}')
print(f'Mahalanobis RGB seul: 0.6684')
print(f'Gain multimodal     : {best_auc_multi - 0.6684:+.4f}')
```

```python
# Subsample train pour accélérer
# Au lieu de tous les recordings → prendre 20 recordings

import numpy as np
import torch

print('=== VERSION RAPIDE ===')

# Utiliser hidden states déjà calculés
# train_hidden est déjà en mémoire ✅
# Pas besoin de re-inférer

train_bigru_preds_fast = []
with torch.no_grad():
    for rec, info in list({**stride_map_train, 
                           **stride_map_test}.items())[:20]:  # 20 seulement
        path = rec_to_file_train.get(rec, rec_to_file_test.get(rec))
        if path is None:
            continue
        arr = np.load(path).astype(np.float32)
        arr = scaler.transform(arr)
        arr = np.clip(arr, -5.0, 5.0)
        n   = arr.shape[0]
        lp  = np.zeros((n, NUM_CLASSES))
        cnt = np.zeros(n)
        for start in range(0, n-32+1, 128):  # stride=128 au lieu de 32
            end   = min(start+256, n)
            chunk = torch.tensor(arr[start:end]).unsqueeze(0)
            logits, _, _, _ = model(chunk)
            l = torch.log_softmax(logits, dim=-1).squeeze(0).numpy()
            T = l.shape[0]
            lp[start:start+T] += l
            cnt[start:start+T] += 1
        valid = cnt > 0
        lp[valid] /= cnt[valid, None]
        train_bigru_preds_fast.extend(np.argmax(lp[valid], axis=1))

print(f'Train preds : {len(train_bigru_preds_fast)}')
```

```python
# train_hidden est DÉJÀ calculé ✅
# pas besoin de re-inférer

# Simuler bigru_preds depuis les logprobs déjà sauvegardés
# On recalcule seulement sur val (déjà fait aussi)

print('=== FEATURES CONTEXTUELLES RAPIDE ===')

# Pour train : utiliser Mahalanobis distance comme proxy
# de la classe (plus proche de prototype CRITIQUE = classe 1)
# → pas besoin de re-inférer BiGRU sur train

# Construire features enrichies simple :
# PCA64 + distance relative aux 2 prototypes

# Prototypes par classe depuis train
X_nor_pca = pca64_final.transform(
    ss.transform(train_hidden[train_error==0]))
X_err_pca = pca64_final.transform(
    ss.transform(train_hidden[train_error==1]))

proto_nor = X_nor_pca.mean(axis=0)
proto_err = X_err_pca.mean(axis=0)

var_nor = X_nor_pca.var(axis=0) + 1e-4
var_err = X_err_pca.var(axis=0) + 1e-4
inv_nor = np.diag(1.0 / var_nor)
inv_err = np.diag(1.0 / var_err)

# Distance à prototype normal ET erreur pour val
print('Calcul distances val...')
dist_to_normal = np.array([
    np.sqrt(np.maximum((v-proto_nor) @ inv_nor @ (v-proto_nor), 0))
    for v in Xval_pca
])
dist_to_error = np.array([
    np.sqrt(np.maximum((v-proto_err) @ inv_err @ (v-proto_err), 0))
    for v in Xval_pca
])

# Ratio = dist_normal / dist_error
# → élevé = plus proche erreur ✅
ratio_score = dist_to_normal / (dist_to_error + 1e-8)
auc_ratio   = roc_auc_score(val_error, ratio_score)
print(f'AUC ratio dist_normal/dist_error : {auc_ratio:.4f}')

# Combiner avec semi-supervisé
ratio_n   = norm(ratio_score[:len(val_error)])
semi_n_f  = norm(scores_final)

for w in [0.3, 0.4, 0.5, 0.6, 0.7]:
    combo = w * semi_n_f + (1-w) * ratio_n
    auc   = roc_auc_score(val_error, combo)
    flag  = '✅' if auc > 0.83 else ''
    print(f'  Semi={w:.1f} Ratio={1-w:.1f} : {auc:.4f} {flag}')

# Lissage W=20
from scipy.ndimage import uniform_filter1d
best_combo = 0.5 * semi_n_f + 0.5 * ratio_n
best_smooth = uniform_filter1d(best_combo[:len(val_error)], size=20)
auc_final_smooth = roc_auc_score(val_error, best_smooth)

fpr_f, tpr_f, _ = roc_curve(val_error, best_smooth)
print(f'\nAUC final lissé W=20  : {auc_final_smooth:.4f}')
print(f'TPR@FPR=5%            : {tpr_f[np.argmin(np.abs(fpr_f-0.05))]:.4f}')
print(f'TPR@FPR=10%           : {tpr_f[np.argmin(np.abs(fpr_f-0.10))]:.4f}')
print(f'TPR@FPR=20%           : {tpr_f[np.argmin(np.abs(fpr_f-0.20))]:.4f}')
```

```python
import os, re, pickle, glob
import numpy as np
import pandas as pd
import torch
from sklearn.preprocessing import StandardScaler as SS
from sklearn.decomposition import PCA
from sklearn.metrics import roc_auc_score, roc_curve
from scipy.ndimage import uniform_filter1d

# ── Chemins ───────────────────────────────────────────────
SCORES_DIR   = '/kaggle/input/datasets/hibabou/project-final-output/scores/'
MODAL_VAL    = '/kaggle/input/datasets/hibabou/multimodal/multimodal_val/'
MODAL_TRAIN  = '/kaggle/input/datasets/hibabou/multimodal/multimodal_train_test/'
PSR_VAL_DIR  = '/kaggle/input/datasets/hibabou/multimodal/psr_labels_val/'
PREP_DIR     = '/kaggle/input/datasets/hibabou/prepbigru/kaggle/working/'

# ── Charger scores déjà calculés ─────────────────────────
print('=== CHARGEMENT SCORES ===')
val_error       = np.load(f'{SCORES_DIR}val_error.npy')
scores_final    = np.load(f'{SCORES_DIR}scores_final.npy')
best_smooth     = np.load(f'{SCORES_DIR}best_smooth.npy')
maha_final      = np.load(f'{SCORES_DIR}maha_final.npy')
maha_gaze       = np.load(f'{SCORES_DIR}maha_gaze.npy')
maha_hands      = np.load(f'{SCORES_DIR}maha_hands.npy')
ratio_score     = np.load(f'{SCORES_DIR}ratio_score.npy')
val_bigru_preds = np.load(f'{SCORES_DIR}val_bigru_preds.npy')

print(f'val_error    : {val_error.shape} | erreurs={val_error.sum()}')
print(f'scores_final : {scores_final.shape}')
print(f'best_smooth  : {best_smooth.shape}')
print(f'AUC actuel   : {roc_auc_score(val_error, best_smooth):.4f}')

def norm(x): return (x - x.min()) / (x.max() - x.min() + 1e-8)

# ── Charger stride_map val ────────────────────────────────
with open(f'{PREP_DIR}stride_map_val.pkl', 'rb') as f:
    stride_map_val = pickle.load(f)
with open(f'{PREP_DIR}scaler.pkl', 'rb') as f:
    scaler = pickle.load(f)

print(f'Stride map val : {len(stride_map_val)} recordings ✅')

# ── Charger PSR val errors ────────────────────────────────
all_psr = []
for rec_dir in os.listdir(PSR_VAL_DIR):
    psr_path = os.path.join(PSR_VAL_DIR, rec_dir, 'PSR_labels_with_errors.csv')
    if os.path.exists(psr_path):
        df = pd.read_csv(psr_path, header=None,
                         names=['frame','step_id','description'])
        df['recording'] = rec_dir
        all_psr.append(df)

psr_val_df = pd.concat(all_psr, ignore_index=True)
psr_val_df['is_error']  = psr_val_df['description'].str.contains(
    'Incorrect|Remove', case=False, na=False).astype(int)
psr_val_df['frame_num'] = psr_val_df['frame'].str.replace('.jpg','').astype(int)

from collections import defaultdict
error_frames_val = defaultdict(list)
for _, row in psr_val_df[psr_val_df['is_error']==1].iterrows():
    error_frames_val[row['recording']].append(row['frame_num'])

print(f'PSR val erreurs : {psr_val_df["is_error"].sum()} ✅')
print('✅ Tout chargé — prêt pour pose.csv')
```

```python
# ── Charger pose.csv et aligner avec stride_map ───────────
print('=== CHARGEMENT POSE.CSV ===')

def load_pose_features(rec, stride_map, modal_dir, clip_val=5.0):
    """
    Charge pose.csv pour un recording
    9 features : orientation tête + épaules + bras
    Aligne avec stride_map → features par feature-frame
    """
    info = stride_map[rec]
    n    = info['n_feats']

    # Chercher pose.csv
    pose_path = os.path.join(modal_dir, rec, 'pose.csv')
    pose_feat = np.zeros((n, 9), dtype=np.float32)

    if os.path.exists(pose_path):
        df_p = pd.read_csv(pose_path, header=None)
        df_p.columns = ['frame'] + [f'p{i}' for i in range(df_p.shape[1]-1)]
        df_p['frame_num'] = df_p['frame'].str.replace('.jpg','').astype(int)

        for _, row in df_p.iterrows():
            feat_idx = round((row['frame_num'] - info['f_min']) / info['stride'])
            feat_idx = max(0, min(feat_idx, n-1))
            pose_feat[feat_idx] = row.iloc[1:10].values.astype(np.float32)

    return pose_feat

# ── Construire features pose train ───────────────────────
print('Construction pose train...')
X_pose_train_list = []

# Charger stride_map train+test
with open(f'{PREP_DIR}stride_map_train.pkl', 'rb') as f:
    stride_map_train = pickle.load(f)

# Stride map test
TRAIN_FEAT_DIR = '/kaggle/input/datasets/hibabou/industrial/data/train/'
TEST_FEAT_DIR  = '/kaggle/input/datasets/hibabou/featuressupp/sf_features_test/'

rec_to_file_train = {}
for f in os.listdir(TRAIN_FEAT_DIR):
    if f.endswith('.npy'):
        rec = f.replace('_features.npy', '')
        rec_to_file_train[rec] = os.path.join(TRAIN_FEAT_DIR, f)

rec_to_file_test = {}
for f in os.listdir(TEST_FEAT_DIR):
    if f.endswith('.npy'):
        rec = re.sub(r'^test_p\d+_', '', f).replace('_features.npy', '')
        rec_to_file_test[rec] = os.path.join(TEST_FEAT_DIR, f)

# Charger labels CSV
col_names = ['recording','action_id','action_name','frame_start','frame_end']
import urllib.request
os.system('curl -L -o /kaggle/working/labels.zip "https://data.4tu.nl/file/b008dd74-020d-4ea4-a8ba-7bb60769d224/778d7d5f-6631-4cb6-9451-f88c574c7991"')
os.system('unzip -o /kaggle/working/labels.zip -d /kaggle/working/labels/')

train_df = pd.read_csv('/kaggle/working/labels/train.csv', header=None, names=col_names)
test_df  = pd.read_csv('/kaggle/working/labels/test.csv',  header=None, names=col_names)
for df in [train_df, test_df]:
    df['frame_start'] = df['frame_start'].str.replace('.jpg','').astype(int)
    df['frame_end']   = df['frame_end'].str.replace('.jpg','').astype(int)

stride_map_test = {}
for rec, path in rec_to_file_test.items():
    sub = test_df[test_df['recording'] == rec]
    if len(sub) == 0:
        continue
    arr    = np.load(path)
    f_min  = sub['frame_start'].min()
    f_max  = sub['frame_end'].max()
    n_feat = arr.shape[0]
    total  = f_max - f_min
    stride = round((total - 32) / (n_feat - 1)) if n_feat > 1 else 1
    stride = max(1, stride)
    stride_map_test[rec] = {'stride': stride, 'f_min': f_min, 'n_feats': n_feat}

all_train_stride = {**stride_map_train, **stride_map_test}

# PSR train errors
PSR_TRAIN_DIR = '/kaggle/input/datasets/hibabou/psr-labels/kaggle/working/psr_labels/'
all_psr_train = []
for rec_dir in os.listdir(PSR_TRAIN_DIR):
    psr_path = os.path.join(PSR_TRAIN_DIR, rec_dir, 'PSR_labels_with_errors.csv')
    if os.path.exists(psr_path):
        df = pd.read_csv(psr_path, header=None,
                         names=['frame','step_id','description'])
        df['recording'] = rec_dir
        all_psr_train.append(df)

psr_train_df = pd.concat(all_psr_train, ignore_index=True)
psr_train_df['is_error']  = psr_train_df['description'].str.contains(
    'Incorrect|Remove', case=False, na=False).astype(int)
psr_train_df['frame_num'] = psr_train_df['frame'].str.replace('.jpg','').astype(int)

error_frames_train = defaultdict(list)
for _, row in psr_train_df[psr_train_df['is_error']==1].iterrows():
    error_frames_train[row['recording']].append(row['frame_num'])

# Construire pose train
for rec, info in all_train_stride.items():
    pose_f = load_pose_features(rec, all_train_stride, MODAL_TRAIN)
    n      = info['n_feats']

    # Masquer erreurs
    err_mask = np.zeros(n, dtype=bool)
    if rec in error_frames_train:
        for ef in error_frames_train[rec]:
            ei = round((ef - info['f_min']) / info['stride'])
            ei = max(0, min(ei, n-1))
            err_mask[max(0,ei-15):min(n,ei+15)] = True

    X_pose_train_list.append(pose_f[~err_mask])

X_pose_train = np.vstack(X_pose_train_list)

# Construire pose val
X_pose_val_list = []
for rec, info in stride_map_val.items():
    pose_f = load_pose_features(rec, stride_map_val, MODAL_VAL)
    X_pose_val_list.append(pose_f)
X_pose_val = np.vstack(X_pose_val_list)

print(f'Pose train : {X_pose_train.shape}')
print(f'Pose val   : {X_pose_val.shape}')

# ── Mahalanobis sur pose ──────────────────────────────────
ss_pose = SS()
X_pose_tr_n = ss_pose.fit_transform(X_pose_train)
X_pose_vl_n = ss_pose.transform(X_pose_val)

proto_pose = X_pose_tr_n.mean(axis=0)
var_pose   = X_pose_tr_n.var(axis=0) + 1e-4
inv_pose   = np.diag(1.0 / var_pose)

maha_pose = np.array([
    np.sqrt(np.maximum((v-proto_pose) @ inv_pose @ (v-proto_pose), 0))
    for v in X_pose_vl_n
])[:len(val_error)]

auc_pose = roc_auc_score(val_error, maha_pose)
print(f'\nAUC Pose seul : {auc_pose:.4f}')

# ── Combinaison RGB + Gaze + Hands + Pose ────────────────
print('\n=== COMBINAISON RGB + GAZE + HANDS + POSE ===')

maha_rgb_n   = norm(maha_final)
maha_gaze_n  = norm(maha_gaze)
maha_hands_n = norm(maha_hands)
maha_pose_n  = norm(maha_pose)
ratio_n      = norm(ratio_score[:len(val_error)])
semi_n       = norm(scores_final)

best_auc_full = 0
best_config   = None

for w_semi in [0.4, 0.5]:
    for w_ratio in [0.2, 0.3]:
        for w_pose in [0.1, 0.2]:
            w_rest = 1 - w_semi - w_ratio - w_pose
            if w_rest < 0:
                continue
            combo = (w_semi  * semi_n +
                     w_ratio * ratio_n +
                     w_pose  * maha_pose_n +
                     w_rest  * maha_rgb_n)
            combo_smooth = uniform_filter1d(combo, size=20)
            auc = roc_auc_score(val_error, combo_smooth)
            if auc > best_auc_full:
                best_auc_full = auc
                best_config   = (w_semi, w_ratio, w_pose, w_rest)
            print(f'Semi={w_semi} Ratio={w_ratio} Pose={w_pose} '
                  f'RGB={w_rest:.1f} : {auc:.4f}')

print(f'\nMeilleur avec pose : {best_auc_full:.4f}')
print(f'Config             : {best_config}')
print(f'Sans pose          : 0.8530')
print(f'Gain               : {best_auc_full - 0.8530:+.4f}')
```

```python
# Pose inversé
maha_pose_inv_n = norm(-maha_pose)
auc_pose_inv = roc_auc_score(val_error, maha_pose_inv_n)
print(f'AUC Pose inversé : {auc_pose_inv:.4f}')

# Combinaison meilleure sans pose
best_no_pose = uniform_filter1d(
    0.5*semi_n + 0.3*ratio_n + 0.2*maha_rgb_n, size=20)
auc_no_pose = roc_auc_score(val_error, best_no_pose)

# Avec pose inversé
best_with_pose_inv = uniform_filter1d(
    0.5*semi_n + 0.3*ratio_n + 0.1*maha_rgb_n + 0.1*maha_pose_inv_n,
    size=20)
auc_with_pose_inv = roc_auc_score(val_error, best_with_pose_inv)

print(f'AUC sans pose      : {auc_no_pose:.4f}')
print(f'AUC pose inversé   : {auc_with_pose_inv:.4f}')

# TPR à différents FPR
from sklearn.metrics import roc_curve
fpr, tpr, _ = roc_curve(val_error, best_no_pose)
print(f'\n=== RÉSULTAT FINAL ===')
print(f'AUC               : {auc_no_pose:.4f}')
print(f'TPR @ FPR=5%      : {tpr[np.argmin(np.abs(fpr-0.05))]:.4f}')
print(f'TPR @ FPR=10%     : {tpr[np.argmin(np.abs(fpr-0.10))]:.4f}')
print(f'TPR @ FPR=20%     : {tpr[np.argmin(np.abs(fpr-0.20))]:.4f}')
```

```python
# ── Seuils adaptatifs par sévérité ───────────────────────
from sklearn.metrics import roc_curve
import numpy as np

fpr, tpr, thresholds = roc_curve(val_error, best_no_pose)

print('=== SEUILS COBOT PAR SÉVÉRITÉ ===\n')

# Trouver seuils optimaux
targets = {
    'STOP    (FPR=2%)' : 0.02,
    'PAUSE   (FPR=5%)' : 0.05,
    'MONITOR (FPR=10%)': 0.10,
    'WATCH   (FPR=20%)': 0.20,
}

seuils = {}
for label, target_fpr in targets.items():
    idx   = np.argmin(np.abs(fpr - target_fpr))
    thresh = thresholds[idx]
    tpr_v  = tpr[idx]
    fpr_v  = fpr[idx]
    seuils[label] = thresh
    print(f'{label} : threshold={thresh:.3f} | '
          f'TPR={tpr_v:.3f} | FPR={fpr_v:.3f}')

# ── Appliquer sur val et compter par recording ────────────
print('\n=== DÉTECTION PAR RECORDING ===')
print(f'{"Recording":>15} {"Erreurs":>8} {"STOP":>6} '
      f'{"PAUSE":>7} {"MONITOR":>8} {"WATCH":>6} {"Raté":>6}')
print('-' * 65)

frame_idx = 0
total_stop = total_pause = total_monitor = total_watch = total_missed = 0
total_errors = 0

for rec, info in stride_map_val.items():
    from collections import defaultdict
    lp_valid = np.ones(info['n_feats'], dtype=bool)
    n = lp_valid.sum()

    scores_rec = best_no_pose[frame_idx:frame_idx+n]

    # Labels erreur
    err_mask = np.zeros(n)
    for ef in error_frames_val.get(rec, []):
        ei = round((ef - info['f_min']) / info['stride'])
        ei = max(0, min(ei, n-1))
        err_mask[max(0,ei-15):min(n,ei+15)] = 1

    n_errors = int(err_mask.sum())
    if n_errors == 0:
        frame_idx += n
        continue

    # Classifier chaque frame erreur
    err_indices = np.where(err_mask == 1)[0]
    err_scores  = scores_rec[err_indices]

    t_stop    = seuils['STOP    (FPR=2%)']
    t_pause   = seuils['PAUSE   (FPR=5%)']
    t_monitor = seuils['MONITOR (FPR=10%)']
    t_watch   = seuils['WATCH   (FPR=20%)']

    n_stop    = (err_scores >= t_stop).sum()
    n_pause   = ((err_scores >= t_pause) & (err_scores < t_stop)).sum()
    n_monitor = ((err_scores >= t_monitor) & (err_scores < t_pause)).sum()
    n_watch   = ((err_scores >= t_watch) & (err_scores < t_monitor)).sum()
    n_missed  = (err_scores < t_watch).sum()

    total_stop    += n_stop
    total_pause   += n_pause
    total_monitor += n_monitor
    total_watch   += n_watch
    total_missed  += n_missed
    total_errors  += n_errors

    print(f'{rec:>15} {n_errors:>8} {n_stop:>6} '
          f'{n_pause:>7} {n_monitor:>8} {n_watch:>6} {n_missed:>6}')

    frame_idx += n

print('-' * 65)
print(f'{"TOTAL":>15} {total_errors:>8} {total_stop:>6} '
      f'{total_pause:>7} {total_monitor:>8} {total_watch:>6} {total_missed:>6}')

print(f'\n=== RÉSUMÉ ===')
detected = total_stop + total_pause + total_monitor + total_watch
print(f'Détectés (WATCH+)  : {detected}/{total_errors} '
      f'({detected/total_errors:.1%})')
print(f'STOP + PAUSE       : {total_stop+total_pause}/{total_errors} '
      f'({(total_stop+total_pause)/total_errors:.1%})')
print(f'Au moins MONITOR   : {total_stop+total_pause+total_monitor}/{total_errors} '
      f'({(total_stop+total_pause+total_monitor)/total_errors:.1%})')
print(f'Complètement ratés : {total_missed}/{total_errors} '
      f'({total_missed/total_errors:.1%})')
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import numpy as np

fig, axes = plt.subplots(4, 3, figsize=(18, 14))
axes = axes.flatten()

frame_idx = 0
plot_idx  = 0

for rec, info in stride_map_val.items():
    n = info['n_feats']
    scores_rec = best_no_pose[frame_idx:frame_idx+n]

    err_mask = np.zeros(n)
    for ef in error_frames_val.get(rec, []):
        ei = round((ef - info['f_min']) / info['stride'])
        ei = max(0, min(ei, n-1))
        err_mask[max(0,ei-15):min(n,ei+15)] = 1

    if err_mask.sum() == 0:
        frame_idx += n
        continue

    frames = np.arange(n)
    ax = axes[plot_idx]

    # Score
    ax.plot(frames, scores_rec, color='steelblue',
            linewidth=1, alpha=0.7, label='Anomaly score')

    # Zones colorées par sévérité
    t_stop    = seuils['STOP    (FPR=2%)']
    t_pause   = seuils['PAUSE   (FPR=5%)']
    t_monitor = seuils['MONITOR (FPR=10%)']
    t_watch   = seuils['WATCH   (FPR=20%)']

    ax.axhline(t_stop,    color='red',    linestyle='--', linewidth=1.2, alpha=0.8)
    ax.axhline(t_pause,   color='orange', linestyle='--', linewidth=1.2, alpha=0.8)
    ax.axhline(t_monitor, color='gold',   linestyle='--', linewidth=1.2, alpha=0.8)
    ax.axhline(t_watch,   color='green',  linestyle='--', linewidth=1.2, alpha=0.8)

    # Zones erreurs PSR annotées
    for i, e in enumerate(err_mask):
        if e:
            ax.axvspan(i-0.5, i+0.5, color='red', alpha=0.25)

    # Colorer les détections par niveau
    for i in range(n):
        if scores_rec[i] >= t_stop:
            ax.axvspan(i-0.5, i+0.5, color='red',    alpha=0.08)
        elif scores_rec[i] >= t_pause:
            ax.axvspan(i-0.5, i+0.5, color='orange', alpha=0.08)
        elif scores_rec[i] >= t_monitor:
            ax.axvspan(i-0.5, i+0.5, color='gold',   alpha=0.08)
        elif scores_rec[i] >= t_watch:
            ax.axvspan(i-0.5, i+0.5, color='green',  alpha=0.08)

    # Stats
    err_idx    = np.where(err_mask==1)[0]
    err_sc     = scores_rec[err_idx]
    n_stop     = (err_sc >= t_stop).sum()
    n_pause    = ((err_sc >= t_pause) & (err_sc < t_stop)).sum()
    n_monitor  = ((err_sc >= t_monitor) & (err_sc < t_pause)).sum()
    n_watch    = ((err_sc >= t_watch) & (err_sc < t_monitor)).sum()
    n_missed   = (err_sc < t_watch).sum()
    n_errors   = int(err_mask.sum())

    ax.set_title(
        f'{rec}\n'
        f'STOP={n_stop} PAUSE={n_pause} MON={n_monitor} '
        f'WATCH={n_watch} MISS={n_missed}/{n_errors}',
        fontsize=8
    )
    ax.set_ylim(0, 1)
    ax.set_xlabel('Frame', fontsize=7)
    ax.grid(True, alpha=0.2)

    frame_idx += n
    plot_idx  += 1

# Légende
legend_patches = [
    mpatches.Patch(color='red',    alpha=0.6, label=f'STOP    (>{t_stop:.2f})'),
    mpatches.Patch(color='orange', alpha=0.6, label=f'PAUSE   (>{t_pause:.2f})'),
    mpatches.Patch(color='gold',   alpha=0.6, label=f'MONITOR (>{t_monitor:.2f})'),
    mpatches.Patch(color='green',  alpha=0.6, label=f'WATCH   (>{t_watch:.2f})'),
    mpatches.Patch(color='red',    alpha=0.25, label='PSR Error zone'),
]
fig.legend(handles=legend_patches, loc='lower center',
           ncol=5, fontsize=9, bbox_to_anchor=(0.5, -0.02))

plt.suptitle(
    'Cobot Decision Engine — Error Detection by Severity\n'
    f'AUC=0.853 | Detected=75.2% | STOP+PAUSE=53.1% | Missed=24.8%',
    fontsize=12, fontweight='bold'
)
plt.tight_layout()
plt.savefig('/kaggle/working/cobot_severity_detection.png',
            dpi=150, bbox_inches='tight')
plt.show()
print('✅ Sauvegardé')
```
